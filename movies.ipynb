```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# download once
# nltk.download('stopwords')
# nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):

    # lowercase
    text = str(text).lower()

    # remove punctuations
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # tokenize
    words = text.split()

    # remove stopwords
    words = [word for word in words if word not in stop_words]

    # lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title, n=10):

    if title not in indices:
        return ["Movie not found"]

    idx = indices[title]

    # compare one movie with all movies
    sim_score = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    # sort similarity scores
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    return df['title'].iloc[similar_idx]
```

```python
from rapidfuzz import process
```

```python
def find_closest_match(query):

    # movie titles list
    movie_titles = df["title"].tolist()

    # genres list
    all_genres = set(" ".join(df["genres"].fillna("")).lower().split())

    # best movie match
    movie_match = process.extractOne(query, movie_titles)

    # best genre match
    genre_match = process.extractOne(query.lower(), list(all_genres))

    return movie_match, genre_match
```

```python
def recommend_smart(query, n=10):

    query = query.strip()

    # lowercase movie matching
    lower_titles = df["title"].str.lower()

    # 1️⃣ Exact movie match (case insensitive)
    if query.lower() in lower_titles.values:
        real_title = df[df["title"].str.lower()==query.lower()]["title"].iloc[0]
        return recommend_movie(real_title, n)

    # 2️⃣ Exact genre match
    if df["genres"].str.lower().str.contains(query.lower(), na=False).any():
        return recommend_by_genre(query, n)

    # 3️⃣ Fuzzy matching (typo correction)
    movie_match, genre_match = find_closest_match(query)

    # similarity score check
    if movie_match and movie_match[1] > 70:
        return recommend_movie(movie_match[0], n)

    if genre_match and genre_match[1] > 70:
        return recommend_by_genre(genre_match[0], n)

    return ["Movie or Genre not found"]
```

```python
# ===============================
# 7️⃣ FUZZY MOVIE MATCH (TYPO FIX)
# ===============================
def get_best_movie_match(query):

    movie_titles = df["title"].str.lower().tolist()

    match = process.extractOne(
        query.lower(),
        movie_titles,
        score_cutoff=65   # stable threshold
    )

    if match:
        real_title = df[
            df["title"].str.lower() == match[0]
        ]["title"].iloc[0]

        return real_title

    return None


# ===============================
# 8️⃣ FINAL SMART RECOMMENDER
# ===============================
def recommend(query, n=10):

    query = query.strip()

    # 1️⃣ Case-insensitive exact movie match
    lower_titles = df["title"].str.lower()

    if query.lower() in lower_titles.values:
        real_title = df[
            df["title"].str.lower() == query.lower()
        ]["title"].iloc[0]

        print("Showing results for movie:", real_title)
        return recommend_movie(real_title, n)

    # 2️⃣ Genre match
    if df["genres"].str.lower().str.contains(query.lower(), na=False).any():
        print("Showing top movies in genre:", query)
        return recommend_by_genre(query, n)

    # 3️⃣ Fuzzy movie match (typo correction)
    best_movie = get_best_movie_match(query)

    if best_movie:
        print("Showing results for:", best_movie)
        return recommend_movie(best_movie, n)

    return ["Movie or Genre not found"]
```
